```python
ab_mags = np.arange(22, 29, 0.5)
times = calc_snr_AB(source_AB_mags=ab_mags, snr=5, darkcurrent=0)
snrs = dict.fromkeys(times.keys())
for band, time in times.items():
    snrs[band] = calc_snr_AB(
        source_AB_mags=ab_mags, t=time, passbands=band, darkcurrent=0
    )
display(pd.DataFrame(times, index=ab_mags))
display(pd.DataFrame(snrs, index=ab_mags))
```

```python
NPIX = 3
INCLUDE_GEOCORONAL = True
ab_mags = np.arange(22, 29, 0.5)
times = calc_snr_AB(source_AB_mags=ab_mags, snr=5, darkcurrent=0.01, npix=NPIX, include_geocoronal=INCLUDE_GEOCORONAL)
snrs = dict.fromkeys(times.keys())
for band, time in times.items():
    snrs[band] = calc_snr_AB(
        source_AB_mags=ab_mags,
        t=time,
        passbands=band,
        darkcurrent=0.01,
        npix=NPIX,
        include_geocoronal=INCLUDE_GEOCORONAL,
    )
display(pd.DataFrame(times, index=ab_mags))
display(pd.DataFrame(snrs, index=ab_mags))
```
